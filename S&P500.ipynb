```python
#Importing libraries
import pandas as pd
import networkx as nx
```

```python
#Reading CSV file
df = pd.read_csv('s&p500_companies.csv')
df.head()
```

```python
#Dropping columns
new_df = df.drop(['Exchange', 'Shortname', 'Longname', 'Industry', 'Currentprice', 'Ebitda', 'Revenuegrowth', 
                 'City', 'State', 'Country', 'Fulltimeemployees', 'Longbusinesssummary', 'Marketcap'],
                 axis=1)
new_df.head()
```

```python
#Saving a graph from Dataframe
g = nx.from_pandas_edgelist(new_df, source = 'Symbol', target = 'Sector', edge_attr = 'Weight')
```

```python
#Saving the graph
nx.write_graphml(g, 's&p4.graphml')
```

```python
#Finding degree of centrality for each symbol and saving in a dictionary
centrality_dict = nx.degree_centrality(g)
#Developing a sorted Dataframe
sorted_df = pd.DataFrame(sorted(centrality_dict.items(), key = lambda x: x[1], reverse = True), columns=['Sector', 'Degree Centrality'])
sorted_df[: 5]
```
